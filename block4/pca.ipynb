# Метод главных компонент

Метод главных компонент (PCA) - это статистический метод уменьшения размерности данных. 

Его суть заключается в том, чтобы найти новые оси (главные компоненты), вдоль которых данные наиболее изменчивы. Затем данные проецируются на эти новые оси, что позволяет представить их в пространстве с меньшим количеством измерений. Главные компоненты являются ортогональными и упорядочены по убыванию объясненной дисперсии. 

Этот метод широко используется для визуализации данных, снижения шума, выделения основных закономерностей и ускорения алгоритмов обучения.

# Лабораторная работа

```python
# Используемые библиотеки

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# Загрузка датасета

iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

```python
# Применение метода главных компонент

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Target'] = y

plt.figure(figsize=(10, 6))
targets = np.unique(y)
colors = ['r', 'g', 'b']
for target, color in zip(targets, colors):
    indices_to_keep = pca_df['Target'] == target
    plt.scatter(pca_df.loc[indices_to_keep, 'PC1'], pca_df.loc[indices_to_keep, 'PC2'], c=color, s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.legend(iris.target_names)
plt.grid()
plt.show()
```
