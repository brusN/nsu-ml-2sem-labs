# Линейная регрессия

Линейная регрессия в машинном обучении — это метод статистического моделирования, который используется для анализа и предсказания отношений между непрерывными зависимыми переменными и одной или несколькими независимыми переменными. 

В сущности, линейная регрессия строит линейную модель, которая аппроксимирует зависимость между независимыми переменными (предикторами) и зависимой переменной (целевой переменной). 

Цель состоит в том, чтобы найти наилучшую линейную функцию, которая наиболее точно предсказывает значение зависимой переменной на основе значений независимых переменных.

# Лабораторная работа

```python
# Используемые библиотеки

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
# Генерируем датасет

X, y = make_regression(n_samples=50, n_features=1, n_informative=1, noise=10, random_state=0)
display(X, y)
```

```python
# Отображаем датасет

fig = plt.figure(figsize=(10, 6))
plt.scatter(X, y)

plt.xlabel('Признак')
plt.ylabel('Таргет')
plt.show()
```

```python
# Обучаем модель

model = LinearRegression()
model.fit(X, y)
model_a = model.coef_[0]
model_b = model.intercept_
```

```python
# Отображаем линию

plt.figure(figsize=(10, 6))

x = np.arange(-3, 3)
model_y_sk = model_a * x + model_b

plt.plot(x, model_y_sk, linewidth=2, c='r', label=f'Уравнение прямой = {model_a:.2f}x + {model_b:.2f}')
plt.scatter(X, y) 
plt.grid()
plt.xlabel('Признак')
plt.ylabel('Таргет')
plt.legend(prop={'size': 16})
plt.show()
```

```python
# Оценка модели

model.score(X, y)
```
